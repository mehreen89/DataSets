```python
import pydotplus # visualize a decision tree
from sklearn.datasets import load_iris
from sklearn import tree
from IPython.display import Image, display
from sklearn.model_selection import train_test_split
from sklearn.tree import export_text
__author__ = "Mayur Kulkarni <mayur.kulkarni@xoriant.com>"
```

```python
def load_data_set():
    """
    Loads the iris data set

    :return:        data set instance
    """
    iris = load_iris()
    return iris
```

```python
def train_model(iris):
    """
    Train decision tree classifier

    :param iris:    iris data set instance
    :return:        classifier instance
    """
    clf = tree.DecisionTreeClassifier()
    clf = clf.fit(iris.data, iris.target)
    return clf
```

```python
iris_data = load_iris()
```

```python
def display_image(clf, iris):
    """
    Displays the decision tree image
    :param clf:     classifier instance
    :param iris:    iris data set instance
    """
    dot_data = tree.export_graphviz(clf, out_file=None,
                                    feature_names=iris.feature_names,
                                    class_names=iris.target_names,
                                    filled=True, rounded=True)

    graph = pydotplus.graph_from_dot_data(dot_data)
    display(Image(data=graph.create_png()))
```

```python
if __name__ == '__main__':
    iris_data = load_iris()
    decision_tree_classifier = train_model(iris_data)
    display_image(clf=decision_tree_classifier, iris=iris_data)
```

In a decision tree visualization, each node and leaf represents specific information about the decision-making process of the model. Here's what is typically represented in each part:

1. **Nodes:**
   - **Splitting Criterion:** Nodes represent decision points based on a feature and a threshold. The splitting criterion indicates the condition for moving down the tree. For example, "sepal length <= 2.45."

   - **Gini Impurity (for classification):** The Gini impurity is a measure of how often a randomly chosen element would be incorrectly labeled. Nodes often display the Gini impurity before and after the split.

   - **Samples:** The number of samples (data points) that reach that node.

   - **Value (for classification):** The distribution of samples among classes at that node. For example, "setosa: 40, versicolor: 10, virginica: 5."

   - **Class (for regression):** For regression tasks, nodes might display the predicted value.

2. **Leaves:**
   - **Value (for classification):** The final distribution of samples among classes in a leaf node. This represents the predicted class for the leaf.

   - **Class (for regression):** The predicted value for regression tasks.

   - **Samples:** The number of samples that reach the leaf.

The idea is that as you traverse the tree from the root to the leaves, each decision point narrows down the possibilities based on certain conditions. The leaves provide the final prediction or classification for the input data. The splitting criteria and impurity measures help to understand the decision boundaries and the importance of different features in making predictions.

# Model training and evaluation

```python
from sklearn.metrics import accuracy_score, classification_report

# Function to evaluate the model
def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred)
    return accuracy, report

# Load the iris dataset
iris_data = load_iris()
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(iris_data.data, iris_data.target, test_size=0.2, random_state=42)
```

```python
# Function to train the model
def train_model(X_train, y_train):
    clf = tree.DecisionTreeClassifier()
    clf = clf.fit(X_train, y_train)
    return clf

# Train the decision tree classifier
decision_tree_classifier = train_model(X_train, y_train)

# Display the decision tree
# display_image(clf=decision_tree_classifier, iris=iris_data)

# Evaluate the model on the test set
accuracy, report = evaluate_model(decision_tree_classifier, X_test, y_test)

# Print the results
print("Accuracy:", accuracy)
print("Classification Report:\n", report)
```

#Improving the model
#a) Cost-Complexity Pruning:

Cost-complexity pruning is a method for trimming a decision tree to reduce its complexity and prevent overfitting. This involves calculating the cost-complexity of each subtree and pruning away subtrees with a high cost-complexity. Cost-Complexity Pruning involves tuning the decision tree by adjusting the cost complexity parameter ccp_alpha. Use cross-validation to find the optimal ccp_alpha.  

```python
from sklearn.model_selection import cross_val_score

# Train the decision tree classifier with Cost-Complexity Pruning
def train_pruned_model(X, y):
    clf = tree.DecisionTreeClassifier(random_state=42)
    path = clf.cost_complexity_pruning_path(X, y)
    ccp_alphas, impurities = path.ccp_alphas[:-1], path.impurities[:-1]

    clfs = []
    for ccp_alpha in ccp_alphas:
        clf = tree.DecisionTreeClassifier(random_state=42, ccp_alpha=ccp_alpha)
        clf.fit(X, y)
        clfs.append(clf)

    scores = [cross_val_score(clf, X, y, cv=5).mean() for clf in clfs]
    optimal_ccp_alpha = ccp_alphas[scores.index(max(scores))]

    # Create and fit a pruned decision tree
    pruned_decision_tree = tree.DecisionTreeClassifier(random_state=42, ccp_alpha=optimal_ccp_alpha)
    pruned_decision_tree.fit(X, y)
    return pruned_decision_tree

# Train the pruned decision tree classifier
pruned_decision_tree_classifier = train_pruned_model(X_train, y_train)

# Display the decision tree
display_image(clf=pruned_decision_tree_classifier, iris=iris_data)

# Evaluate the pruned model on the test set
accuracy, report = evaluate_model(pruned_decision_tree_classifier, X_test, y_test)

# Print the results
print("Accuracy with Pruning:", accuracy)
print("Classification Report with Pruning:\n", report)
```

#b) Bagging or Random Forest:

Bagging involves training multiple instances of the same learning algorithm on different subsets of the training data and combining their predictions. Random Forest is an ensemble method that builds multiple decision trees and merges them together to get a more accurate and stable prediction.

```python
from sklearn.ensemble import RandomForestClassifier

# Train a Random Forest classifier with 100 trees
random_forest_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
random_forest_classifier.fit(X_train, y_train)

# Display the decision tree
display_image(clf=pruned_decision_tree_classifier, iris=iris_data)

# Evaluate the Random Forest model on the test set
accuracy, report = evaluate_model(random_forest_classifier, X_test, y_test)

# Print the results
print("Accuracy with Random Forest:", accuracy)
print("Classification Report with Random Forest:\n", report)
```

#c) Boosting:

Boosting is an ensemble learning method that trains decision trees sequentially, with each tree focusing on the errors made by the previous trees. This helps to correct the mistakes of the earlier trees and improve overall accuracy. One common implementation is AdaBoost.

```python
from sklearn.ensemble import AdaBoostClassifier

# Train an AdaBoost classifier with a decision tree of maximum depth 1 as the base learner and 50 estimators
adaboost_classifier = AdaBoostClassifier(base_estimator=tree.DecisionTreeClassifier(max_depth=1), n_estimators=50, random_state=42)
adaboost_classifier.fit(X_train, y_train)

# Display the decision tree
display_image(clf=pruned_decision_tree_classifier, iris=iris_data)

# Evaluate the AdaBoost model on the test set
accuracy, report = evaluate_model(adaboost_classifier, X_test, y_test)

# Print the results
print("Accuracy with AdaBoost:", accuracy)
print("Classification Report with AdaBoost:\n", report)
```
